## Dataset Overview

> This dataset includes information about individual rides made in the Baywheels bike sharing system covering the greater San Fransisco Bay area in the First 3 months of 2022.
The dataset was obtained from https://s3.amazonaws.com/baywheels-data/index.htmlhttps://s3.amazonaws.com/baywheels-data/index.html and multiple files were joined together to obtain a single dataset which was wrangled to make it fit and clean for analysis.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_2022=pd.read_csv('C:\\Users\\HP\\Downloads\\Clean_Baywheels_2022.csv')
```

## Which Bike Type takes Lesser Time?

The electric bike was by far the fastest bike on average

```python
plt.figure(figsize=[15,9])
sb.stripplot(data=df_2022,x = 'type_of_ride',y='duration_in_minutes' , hue = 'User_type',jitter = 0.35,dodge = True)
plt.legend(loc = 'center left',bbox_to_anchor = (1,0.5),title = 'User Type')
plt.xlabel('Type of Ride')
plt.ylabel('Trip Duration In Minutes')
plt.title('Trip Duration for each type of ride based on User Type');
```

##  Which bike should be purchased if the Baywheels bike sharing system wants to expand its current capacity?

The electric bike should be the obvious choice since it takes less time and its prferred by users while the docked bikes should be reduced since it isnt fancied by users

```python
plt.figure(figsize=[15,8])

sb.countplot(data= df_2022 ,x ='type_of_ride', hue='User_type');
plt.title('Number of Rides for each Bike type',y=1 ,fontweight = 'bold')
plt.xlabel('Type of Bike',fontweight = 'bold')
plt.ylabel('Number of rides',fontweight = 'bold');
```

## Which Month had the Quickest ride time on average?
 
 In the month of March,there were quicker(shorter) rides

```python
plt.figure(figsize=[15,9])
base_color=sb.color_palette()[3]
sb.pointplot(data = df_2022, x='month',y = 'duration_in_minutes', kind = 'Point')
plt.xlabel('Month')
plt.ylabel('Average Duration of Bike Trips in Minutes')
plt.title('Average Duration of Bike Trips each Month');
```

## Is the method at which services is being carried out working?
 
 This is portrayed by a steady increase in the number of rides across the first 3 months of 2022

```python
months = df_2022.month.value_counts()
plt.figure(figsize=[15,9])
base_color = sb.color_palette()[3]
monthorder = df_2022.month.value_counts().index
sb.countplot(data = df_2022 , x = 'month' , color = base_color , order = monthorder);
plt.title('Number of Bike Rides each month',fontweight='bold',y=1);
plt.xlabel('Months',fontweight='bold');
plt.ylabel('Number of Bike rides',fontweight='bold');
```
